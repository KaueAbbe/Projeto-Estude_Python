# Regex
* O que é regex?

É uma forma padronizada de escrever seguimentações de palavras.
Uma aplicação é validação de dado de entrada: CEP, RG, CPF e e-mails.

[link para teste regex](https://regex101.com)

```python
 import re
 re.findall(r"<p>",
            questao_portugues)   # Retorna uma lista com todas expressão regular encontrada num texto
```

```python
questao_portugues
```

```python
re.findall(r"<.*?>",
            questao_portugues)
```

```python
re.sub(r"<.*?>", "T--------E-----S----T--------E", questao_portugues)   #Substitui a regex por algum texto
```

```python
re.search(r"70","12782178217218721827170" )   # retorna o local onde está posicionando um trecho de texto numa frase
```

```python
regex = re.compile(r"70")
regex.search("12782178217218721827170")
```
