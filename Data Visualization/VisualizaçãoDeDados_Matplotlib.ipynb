#<font size = 5 color = 'Blue'>Importando bibliotecas e Dados</font>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  
#Faz gráficos aparecerem de forma automática no Jupyter Notebook
```

```python
url = 'https://raw.githubusercontent.com/KaueAbbe/Projeto-Estude_Python/main/Dados/monitoramento_tempo.json'
df = pd.read_json(url)
df.head()
```

```python
df.info()
```

```python
df['data'] = pd.to_datetime(df['data'])
```

```python
df.info()
```

#<font size = 5 color = 'Blue'>Primeiro Gráfico</font>

```python
# Criar o espaço da figura
fig = plt.figure(figsize = (12,6))
eixo = fig.add_axes([0,0,1,1])  # Escreve qual a porcenta na horizontal e na vertical quero que os eixos estejam

# Criar o plot dentro da figura
eixo.plot(df['data'], df['temperatura'], color = 'slateblue')

#Detalhes das figuras(Títulos, Labels, afins)
eixo.set_title('Temperatura em função da Data', fontsize = 18)
eixo.set_xlabel('Data', fontsize = 14)
eixo.set_ylabel('Temperatura (K)',fontsize = 14)
eixo.legend(['Temperatura'],loc ='best', fontsize = 13)
```

# <font size = 5 color ='blue'>Anotações e Barra</font>

```python
fig = plt.figure(figsize = (12,6))
eixo =fig.add_axes([0,0,1,1])

# Criar o plot dentro da figura
eixo.plot(df['data'], df['temperatura'], color = 'slateblue')


#Detalhes das figuras(Títulos, Labels, afins)
eixo.set_title('Temperatura em função da Data para Maio/2014', fontsize = 18, pad = 20)
eixo.set_xlabel('Data', fontsize = 14)
eixo.set_ylabel('Temperatura (K)',fontsize = 14)
eixo.legend(['Temperatura'],loc ='lower left', fontsize = 13)
eixo.grid(True)

#Adicionando Anotações
eixo.axhline(max(df['temperatura']),color = 'red',linestyle ='--')
eixo.axhline(min(df['temperatura']),color = 'red',linestyle ='--')

x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])
x2 = df['data'][df['temperatura'].idxmax()-7000]
y2 = max(df['temperatura']) - 5
eixo.annotate("Máximo",(x1,y1),fontsize = 20,xytext = (x2,y2),
              arrowprops = dict(facecolor ='k'))

x1 = df['data'][df['temperatura'].idxmin()]
y1 = min(df['temperatura'])
x2 = df['data'][df['temperatura'].idxmin()-7000]
y2 = min(df['temperatura']) + 5
eixo.annotate("Mínimo",(x1,y1),fontsize = 20,xytext = (x2,y2),
              arrowprops = dict(facecolor ='k'))
```

# <font size = 5 color = 'blue'>Gráfico Barras</font>

```python
temperatura_dia_semana = df.groupby('dia_da_semana')['temperatura'].mean()
nome_dias =['Domingo','Segunda','Terça','Quarta','Quinta','Sexta','Sábado']
temperatura_dia_semana = temperatura_dia_semana[nome_dias]
temperatura_dia_semana
```

```python
fig = plt.figure(figsize=(12,6))
eixo =fig.add_axes([0,0,1,1])


cores = ['blue']*7
cores[0] = 'black'
eixo.bar(nome_dias,temperatura_dia_semana, color = cores)



eixo.set_title('Média de temperatura por dia da semana', fontsize = 18)
eixo.set_ylabel('Temperatura (K)', fontsize = 14)
```

# <font color = 'blue'>Gráfico de Dispersão

```python
url = 'https://raw.githubusercontent.com/KaueAbbe/Projeto-Estude_Python/main/Dados/iris.json'
iris = pd.read_json(url)
iris.head()
```

```python
iris['espécie'].unique()
```

```python
fig = plt.figure(figsize=(12,6))
eixo = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa':'red','Iris-versicolor':'Blue','Iris-virginica':'green'}
marcadores = {'Iris-setosa':'x','Iris-versicolor':'o','Iris-virginica':'v'}

for i in iris['espécie'].unique():
  tmp = iris[iris['espécie']== i]
  eixo.scatter(tmp['comprimento_sépala'],tmp['largura_sépala'], color = cores[i], marker = marcadores[i],s =80)


eixo.set_title('Largura em função do comprimento para sépalas', fontsize = 18,pad=10)
eixo.set_xlabel('Comprimento de Sépalas',fontsize = 14)
eixo.set_ylabel('Largura de Sépalas', fontsize = 14)
eixo.tick_params(labelsize = 12)
eixo.legend(cores,fontsize = 15)
```

# <font color = 'blue'>Gráfico Boxplot

```python
fig = plt.figure(figsize=(12,6))
eixo = fig.add_axes([0,0,1,1])

eixo.boxplot(iris.drop('espécie',axis=1).values,patch_artist = True)


eixo.set_title('Boxplot para cada variável', fontsize = 18,pad=10)
eixo.set_xticklabels(iris.drop('espécie',axis=1).columns,fontsize = 14)
eixo.set_ylabel('Distruibuição', fontsize = 14)
eixo.tick_params(labelsize = 12)
```

#<font color ='blue'> Gráfico Histograma</font>

```python
fig = plt.figure(figsize=(12,6))
eixo = fig.add_axes([0,0,1,1])


eixo.hist(iris['comprimento_pétala'],bins =20,density = True)


eixo.set_title('Histograma do comprimento da pétala', fontsize = 18,pad=10)
eixo.set_xlabel('Comprimento da Pétala',fontsize = 14)
eixo.set_ylabel('Densidade', fontsize = 14)
eixo.tick_params(labelsize = 12)
eixo.grid(True)
```

```python
fig = plt.figure(figsize=(12,6))
eixo = fig.add_axes([0,0,1,1])


eixo.hist(iris['comprimento_pétala'],bins =20)



eixo.set_title('Histograma do comprimento da pétala', fontsize = 18,pad=10)
eixo.set_xlabel('Comprimento da Pétala',fontsize = 14)
eixo.set_ylabel('Densidade', fontsize = 14)
eixo.tick_params(labelsize = 12)
eixo.grid(True)

# Adicionando Média e desvio padrão
mu, sigma = iris['comprimento_pétala'].mean(), iris['comprimento_pétala'].std()

eixo.annotate('$\mu = {0:.2f}$\n $\sigma={0:.2f}$'.format(mu,sigma),
              xy = (4.5,20), fontsize = 18)

eixo.axvline(mu,color ='purple',linestyle = '--')
eixo.annotate('Média',xy =(mu-0.5,30), fontsize = 18,color = 'purple')
```
